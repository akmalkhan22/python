# Python Data Types

## Numeric Data types
Python has main three numeric types
### Integer(int)
The example of integer data type are
2,4543, -12
### Float (f)
34.2, 44.9
### Complex number
2+4j

#### Integer

```python
num =34

print("the number is ",num)
print("the type of number is ",type(num))
```

#### FLoat number

```python
float_num = 34.5
print("the float number is {}" .format(float_num))
print("the type of float number is ",type(float_num))
```

#### Complex Number

```python
complex_num = 3 + 4j
x = complex(5, 7)
print("the complex number is ", complex_num)
print(f"the complex number is {x}")
print("the type of complex number is ", type(complex_num))
```

### Strings

```python
Sport = "Cricket"
sentence = "I love to play " + Sport
print(sentence)
```

```python
country = "Pakistan"
print(country[0:])
print(country[0::2])
print(country[1:4])
print(country[-1::-1])
```

### Boolean

```python
python_easy = True

if python_easy:
    print("Python is easy to learn")
else:
    print("Python is hard to learn")
```
